```python
df = pd.read_csv(r"SalinanOnlineRetailData.csv", header=0)
df = df[df['price'].between(0,100,inclusive='right')].reset_index(drop=True)
df
```

```python
df.info()
```

# Binning dengan banyaknya bins tertentu dan panjang masing-masing bin sama

```python
df1 = df.copy()
category = ['Cheap','Medium','Expensive']
df1['bin1'] = pd.cut(df1['price'], 3, labels=category, include_lowest=True)
df1
```

```python
bin1_summary = df1.groupby('bin1', observed=True).agg(
    row_count=('product_code','count'),
    min_price=('price','min'),
    max_price=('price','max')
)
bin1_summary['bin1_interval'] = bin1_summary['max_price'] - bin1_summary['min_price']  
bin1_summary
```

```python
df2 = df1.copy()
bins = np.linspace(min(df['price']), max(df['price']),4)
category = ['Cheap','Medium','Expensive']
df2['bin2'] = pd.cut(df['price'], bins, labels=category, include_lowest=True)
df2
```

```python
bin2_summary = df2.groupby('bin2', observed=True).agg(
    row_count=('product_code','count'),
    min_price=('price','min'),
    max_price=('price','max')
)

bin2_summary['bin2_interval'] = bin2_summary['max_price'] - bin2_summary['min_price'] 
bin2_summary
```
